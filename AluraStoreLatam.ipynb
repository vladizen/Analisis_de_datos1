### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

#1. Análisis de facturación



```python
#Preparando los datos
tienda1['tienda'] = 'Tienda 1'
tienda2['tienda'] = 'Tienda 2'
tienda3['tienda'] = 'Tienda 3'
tienda4['tienda'] = 'Tienda 4'

#Comenzamos con el analisis de facturaciòn de las tiendas
tienda_total1 = tienda1["Precio"].sum()
tienda_total2 = tienda2["Precio"].sum()
tienda_total3 = tienda3["Precio"].sum()
tienda_total4 = tienda4["Precio"].sum()

print(f"La facturación total de la tienda 1 es de ${tienda_total1}")
print(f"La facturación total de la tienda 2 es de ${tienda_total2}")
print(f"La facturación total de la tienda 3 es de ${tienda_total3}")
print(f"La facturación total de la tienda 4 es de ${tienda_total4}")
```

# 2. Ventas por categoría

```python
#Calculamos el total de las ventas por categoria de producto
for i, tienda in enumerate([tienda1, tienda2, tienda3, tienda4], start=1):
    print(f"\nTienda {i} - Total de Ventas por Categoria:")
    print(tienda['Categoría del Producto'].value_counts())
```

# 3. Calificación promedio de la tienda


```python
# Calculamos la calificación promedio de cada tienda
calificacion_promedio1 = tienda1["Calificación"].mean()
calificacion_promedio2 = tienda2["Calificación"].mean()
calificacion_promedio3 = tienda3["Calificación"].mean()
calificacion_promedio4 = tienda4["Calificación"].mean()

print(f"\nLa calificación promedio de la tienda 1 es de {calificacion_promedio1:.2f}")
print(f"La calificación promedio de la tienda 2 es de {calificacion_promedio2:.2f}")
print(f"La calificación promedio de la tienda 3 es de {calificacion_promedio3:.2f}")
print(f"La calificación promedio de la tienda 4 es de {calificacion_promedio4:.2f}")
```

# 4. Productos más y menos vendidos

```python
tiendas = {
            "Tienda 1": tienda1,
            "Tienda 2": tienda2,
            "Tienda 3": tienda3,
            "Tienda 4": tienda4
          }

# Mostrar TOP 3 productos que más y menos dinero generaron por tienda
for nombre, df in tiendas.items():
    df["Precio"] = df["Precio"].astype(float)

    # Agrupar por producto y sumar ingresos
    ingresos = df.groupby("Producto")["Precio"].sum().sort_values(ascending=False)

    top_3 = ingresos.head(3)
    bottom_3 = ingresos.tail(3)

    print(f"\n{nombre} - TOP 3 productos MÁS vendido:")
    for producto, ingreso in top_3.items():
        print(f"  {producto} - Ingresos: ${ingreso:,.2f}")

    print(f"\n{nombre} - TOP 3 productos MENOS vendido:")
    for producto, ingreso in bottom_3.items():
        print(f"  {producto} - Ingresos: ${ingreso:,.2f}")
```

# 5. Envío promedio por tienda

```python
# Calcular el costo de envío promedio por tienda
envio_promedio1 = tienda1["Costo de envío"].mean()
envio_promedio2 = tienda2["Costo de envío"].mean()
envio_promedio3 = tienda3["Costo de envío"].mean()
envio_promedio4 = tienda4["Costo de envío"].mean()

print(f"\nEl costo de envío promedio de la tienda 1 es de ${envio_promedio1:.2f}")
print(f"El costo de envío promedio de la tienda 2 es de ${envio_promedio2:.2f}")
print(f"El costo de envío promedio de la tienda 3 es de ${envio_promedio3:.2f}")
print(f"El costo de envío promedio de la tienda 4 es de ${envio_promedio4:.2f}")
```

#6. Gráficos

```python
import matplotlib.pyplot as plt

ingresos_por_tienda = {}

for nombre, df in tiendas.items():
    df["Precio"] = df["Precio"].astype(float)
    ingresos_por_tienda[nombre] = df["Precio"].sum()

mejor_tienda = max(ingresos_por_tienda, key=ingresos_por_tienda.get)

colores = [
    'orange' if tienda == mejor_tienda else 'teal'
    for tienda in ingresos_por_tienda.keys()
]

plt.figure(figsize=(8, 5))
plt.bar(ingresos_por_tienda.keys(), ingresos_por_tienda.values(), color=colores)
plt.title("Ingresos Totales por Tienda")
plt.ylabel("Ingresos ($)")
plt.xlabel("Tiendas")
plt.grid(axis='y', linestyle='--', alpha=0.6)

valor_destacado = ingresos_por_tienda[mejor_tienda]
plt.text(mejor_tienda, valor_destacado + valor_destacado * 0.01,
         f"${valor_destacado:,.2f}", ha='center', color='black', fontweight='bold')

plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt

promedio_ventas = {}

for nombre, df in tiendas.items():
    df["Precio"] = df["Precio"].astype(float)
    promedio_ventas[nombre] = df["Precio"].mean()

nombres_tiendas = list(promedio_ventas.keys())
valores_promedio = list(promedio_ventas.values())

indice_mejor = valores_promedio.index(max(valores_promedio))
mejor_tienda = nombres_tiendas[indice_mejor]
mejor_valor = valores_promedio[indice_mejor]

plt.figure(figsize=(8, 5))
plt.plot(nombres_tiendas, valores_promedio, marker='o', linestyle='-', color='green', label="Promedio")

plt.plot(mejor_tienda, mejor_valor, marker='o', markersize=10, color='red', label="Mejor promedio")
plt.text(mejor_tienda, mejor_valor + mejor_valor * 0.02, f"${mejor_valor:,.2f}",
         ha='center', color='red', fontweight='bold')

plt.title("Promedio de Venta por Tienda")
plt.xlabel("Tiendas")
plt.ylabel("Promedio de Venta ($)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```

#Análisis del desempeño geográfico

```python
# Combinar todos los dataframes en uno solo
all_stores_df = pd.concat([tienda1, tienda2, tienda3, tienda4], ignore_index=True)
```

```python
import matplotlib.pyplot as plt

# Definir colores para cada tienda
colors = {
    "Tienda 1": "red",
    "Tienda 2": "blue",
    "Tienda 3": "green",
    "Tienda 4": "purple"
}

# Graficar ventas por ubicación para cada tienda
plt.figure(figsize=(12, 8))

for nombre, df in tiendas.items():
    plt.scatter(df["lon"], df["lat"], label=nombre, alpha=0.6, s=20, color=colors[nombre])

plt.title("Distribución geográfica de ventas por tienda")
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Analizar el rendimiento por Lugar de Compra
location_performance = all_stores_df.groupby("Lugar de Compra").agg(
    total_venta=("Precio", "sum"),
    rating=("Calificación", "mean")
).reset_index()

# Mostrar los lugares con mayor ingreso
high_revenue_locations = location_performance.sort_values(by="total_venta", ascending=False).head(10)
print("\nLugares con mayor ingreso:")
display(high_revenue_locations)

# Mostrar los lugares con mayor calificación promedio
high_rating_locations = location_performance.sort_values(by="rating", ascending=False).head(10)
print("\nLugares con mayor calificación promedio:")
display(high_rating_locations)
```
